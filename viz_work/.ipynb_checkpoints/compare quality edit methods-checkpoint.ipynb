```python
#read csv of distribution edit fail counts
path = '../linked_dist_stats/'
dist_edits_df = pd.read_csv(path+'distribution_edit_fails_all.csv') #load edit fails csv
dist_edits_VA_df = pd.read_csv(path + 'distribution_edit_fails_VA.csv')
dist_edits_FHA_df = pd.read_csv(path + 'distribution_edit_fails_FHA.csv')
dist_edits_CONV_df = pd.read_csv(path + 'distribution_edit_fails_CONV.csv')

dist_edits_df.head()
```

```python
dist_edits_df = dist_edits_df[dist_edits_df.index < 25]
fails = plt.figure(figsize=(18,10))
xticks = [x[-3:] for x in dist_edits_df['bin']] #get top number of each bin range

ax = plt.subplot() #instantiate axis object to enable extra functions
x = range(len(dist_edits_df['bin'])) #set x axis range
plt.xticks(x,xticks) #set xtick names

y = dist_edits_df['count'] #set y axis values
y2 = dist_edits_CONV_df['count']
y3 = dist_edits_VA_df['count']
y4 = dist_edits_FHA_df['count'] 

plt.plot(x,y, label='all edit fails')
plt.plot(x,y2, label='CONV fails')
plt.plot(x,y3, label='VA fails')
plt.plot(x,y4, label='FHA fails')

plt.legend(loc='upper right') #turn on legend and set location
plt.ylabel('count of edit fails')
plt.xlabel('income bin')
plt.show(fails)
```
